**Problem 1: Mobile Recharge Validation System**

**Problem Statement**
Create a function to validate a mobile recharge request.

**Rules:**
Recharge amount must be ≥ ₹50

**Valid plans**: 199, 299, 399, 599

User should be allowed to retry if invalid

**Real-Time Use**
Telecom recharge apps
Payment gateways

**Hint**
Use function for validation
Use while loop for retry


```python
#Creating a function to validate the recharge plan with respect to amount and valid plans,
#If any one of the condition fails needs to Retry
def validation(recharge_amount,valid_plans):
  plan = int(input())
  while recharge_amount < 50 or plan not in valid_plans:
    if(recharge_amount < 50):
      print("Invalid Recharge Amount")
      recharge_amount = int(input("Enter Recharge Amount greater than 50: "))
    if(plan not in valid_plans):
      print("Invalid Plan")
      plan = int(input("Enter Valid Plan: "))
  return recharge_amount,plan
recharge_amount = int(input())
valid_plans = [199,299,399,599]
print(f"Valid Recharge Amount,Plan: [{validation(recharge_amount,valid_plans)}]")
```

**Problem 2: Inventory Reorder Alert System**

**Problem Statement**

Given a dictionary of products and stock quantities:

Print Reorder Alert if stock < 15

Print Stock OK otherwise


**Real-Time Use**

Warehouse systems

Retail management


**Hint**

Use function

Use for loop on dictionary


```python
#Creating an alert system, if stock of any item/product becomes less than 15 then alert message need to be send,
#Otherwise Okay for that product
def alert_system(dic):
  for pro,stock in dic.items():
    if(stock<15):
      print(f"Alert!!! Stock of {pro} is {stock}")
    else:
      print(f"{pro}-->{stock} , Okay Efficient stock is there")

n = int(input("Enter number of products: "))
dic = {}
for i in range(n):
  pro = input("Enter Product Name: ")
  stock = int(input("Enter Stock Quantity: "))
  dic[pro] = stock
print(dic)
print(f"{alert_system(dic)}")
```

**Problem 3: Student Result Processing System**

**Problem Statement**

Create a function that:

Accepts a list of student marks

Calculates average marks


Prints Pass if average ≥ 50 else Fail


**Real-Time Use**

Online exam systems

Evaluation portals


**Hint**

Use loop to calculate total


Return result from function




```python
#creating student result processing system function, if average is greater than 50 then Pass otherwise Fail
def student_result_ps(marks):
  total = 0
  avg = 0
  for i in marks:
    total += i
  avg = total/len(marks)
  if(avg>=50):
    return "Pass"
  else:
    return "Fail"
marks = list(map(int,input().split()))
print(f"{student_result_ps(marks)}")
```

**Problem 4: Cab Fare Estimator with Retry Option**

**Problem Statement**
Create a cab fare calculator:
Base fare = ₹50

₹12 per km


Peak hour → 25% extra


Allow retry using while loop


**Real-Time Use**

Ride-hailing apps


Transport systems


**Hint**

Function for fare calculation


While loop for retry


```python
#creating a cab fare calculator, base price is 50
#And in loop we are checking of correct peak hour status, valid distance
def cab_fare(distance,peak_hour):
  base_price = 50
  price = 0
  while(distance<=0 or peak_hour not in ["yes","no"]):
    if(distance<=0):
      print("Invalid Distance")
      distance = int(input("Enter Valid Distance: "))
    if(peak_hour not in ["yes","no"]):
      print("Invalid Peak Hour")
      peak_hour = input("Enter Valid Peak Hour: ")
  price = base_price+distance*12
  if(peak_hour=="yes"):
    price+=0.25*price
  return price
distance = int(input())
peak_hour = input().lower()
print(f"Cab Fare: {cab_fare(distance,peak_hour)}")
```

**Problem 5: Employee Attendance Eligibility Checker**

**Problem Statement**

Create a function that:

Accepts attendance list ("P", "A")

Calculates attendance percentage

Returns Eligible if ≥ 75% else Not Eligible


**Real-Time Use**

HR payroll systems


Internship tracking


**Hint**

Use loop to count present days


Use return statement


```python
#Checking Employee attendance and attendance should be greater than 75%
def attendance_checker(attendance_list):
  count = 0
  for i in attendance_list:
    if(i=="P"):
      count+=1
  if((count/len(attendance_list))*100>=75):
    return "Eligible"
  else:
    return "Not Eligible"
attendance_list = list(map(str,input().upper().split()))
print(attendance_list)
print(f"Attendance Eligibility: {attendance_checker(attendance_list)}")
```

**Problem 6: Password Strength Checker**

**Problem Statement**

Create a function to check password strength.

Rules:
Minimum length = 8

Must contain at least one digit


Must contain at least one special character (@#$)


**Real-Time Use**

Login systems


Security validation


**Hint**

Use loop over string


Use conditions inside function


```python
#creating strength checker, password is strong if and only if length>=8, contain one digit and one special character
def strength_checker(password):
  count_len = 0
  count_digit = 0
  count_special_character = 0
  for i in password:
    count_len+=1
    if(i.isdigit()):
      count_digit+=1
    if(i in ["@","#","$"]):
      count_special_character+=1
  if(count_len>=8 and count_digit>=1 and count_special_character>=1):
    print("Strong Password")
  else:
    print("Create a New Password that contains digits, special chars and length of password>=8")
password = input()
strength_checker(password)
```
