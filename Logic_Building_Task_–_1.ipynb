**2. Pass / Fail Analyzer**

Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks ≥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly

**Real-World Application: Academic evaluation systems**


```python
#Here we are taking the number of students as input --> n
#We are creating the pass_count,fail_count variable to store respective values
#We are also creating a list of n students with there marks
n = int(input())
pass_count = 0
marks = []
for i in range(n):
  val = int(input())
  marks.append(val)
for i in marks:
  if(i >= 50):
    pass_count += 1
fail_count = n - pass_count
print(f"Marks of Student : {marks}")
print("Pass Count:",pass_count)
print("Fail Count:",fail_count)
```

**3. Simple Data Cleaner**

Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list

**Real-World Application: Data preprocessing before analysis**


```python
# Taking n input names from user and performing
n = int(input())
names = []
for i in range(n):
  val = input()
  names.append(val)
print(f"Names before change: {names}")
for i in range(len(names)):
  names[i] = names[i].strip()
  names[i] = names[i].lower()
print(f"Names after change: {names}")
```

**4. Message Length Analyzer**

Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters

**Real-World Application: Text filtering and validation systems**


```python
#Taking n messages from user and printing the message accepted/ not accepted condition..
n = int(input())
messages = []
for i in range(n):
  val = input()
  messages.append(val)
for i in messages:
  if(len(i) > 10):
    print(f"{i}-->Message Accepted , length of message {len(i)}")
  else:
    print(f"{i}-->Message not Accepted , length of message {len(i)}")
```

**5. Error Message Detector**

Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count

**Real-World Application: Monitoring and log analysis systems**

```python
#taking n logs and checking the count of "ERROR"
n = int(input())
logs = []
count = 0
for i in range(n):
  val = input()
  logs.append(val)
for i in logs:
  if(i == "ERROR"):
    count+=1
print(f"Number of ERROR entries are : {count}")
```
