**Problem Statement 1: Employee Performance Bonus Eligibility**

**Description:**
A company evaluates employee performance scores at the end of the year. You are given a dictionary containing employee names and their performance scores.

**Requirements:**

Identify the highest performance score.
Handle ties if multiple employees have the same highest score.
Display all employees eligible for the top performance bonus.

**Input:**
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

**Expected Output:**
Top Performers Eligible for Bonus: Ravi, Kiran (Score: 92)

```python
#Taking n employees input and storeing there name,score in dictionary
#After that calculating the max score and printing bonus eligible employees
n = int(input())
employees={}
for i in range(n):
    name = input()
    score = int(input())
    employees[name]=int(score)
print(employees)
top_performace_score = max(employees.values())
list_of_top_performers = []
for name,score in employees.items():
    if score == top_performace_score:
        list_of_top_performers.append(name)
print("Top Performers Eligible for Bonus: ",", ".join(list_of_top_performers),"(Score: ",top_performace_score,")")
```

**Problem Statement 2: Search Query Keyword Analysis**

**Description:**
An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.

**Requirements:**
Convert the input to lowercase.
Ignore common punctuation.
Count the frequency of each keyword.
Display only keywords searched more than once.

**Input:**
"Buy mobile phone buy phone online"

**Expected Output:**
{'buy': 2, 'phone': 2}


```python
#Taking input query from user and storeing frequecy of each query.
#Before storeing checking that key should not contain any punctuation,
#At last just printing the values if count of frequecy is greater than equall to 2
input_string = input()
input_string = input_string.lower()
list_input_string = input_string.split()
punctuation = [".",",","!","?"]
frequecy_query = {}
for i in list_input_string:
  if i not in punctuation:
    frequecy_query[i] = frequecy_query.get(i,0)+1
ans = {}
for word,count in list(frequecy_query.items()):
  if count>=2:
    ans[word] = count
print(ans)

```

**Problem Statement 3: Sensor Data Validation**

**Description:**
A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.

**Requirements:**
Identify readings that are even numbers (valid readings).
Store them as (hour_index, reading_value) pairs.
Ignore odd readings (invalid readings).

**Input:**
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

**Expected Output:**
Valid Sensor Readings (Hour, Value):
[(1, 4), (3, 8), (4, 10), (5, 12)]

```python
#Taking n inputs and making sensor reading and valid sensor reading list
#in valid sensor reading we are storeing tuple(index,sensor reading) only if sensor reading is EVEN
n = int(input())
sensor_readings = []
valid_sensor_readings = []
for i in range(n):
  val = int(input())
  sensor_readings.append(val)
for i in range(n):
  if(sensor_readings[i]%2==0):
    valid_sensor_readings.append((i,sensor_readings[i]))
print(f"Valid Sensor Readings (Hour, Value): {valid_sensor_readings}")
```

**Problem Statement 4: Email Domain Usage Analysis**

**Description:**
A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.

**Requirements:**
Count how many users belong to each email domain.
Calculate the percentage usage of each domain.

**Input:**
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

**Expected Output:**
gmail.com: 60%
yahoo.com: 40%

```python
#Taking users_count from users
#Storeing emails in emails list and printing percentages of gmail user and yahoo user
users_count = int(input())
emails_list = []
for i in range(users_count):
  email = input()
  emails_list.append(email)
count_gmail = 0
count_yahoo = 0
for i in emails_list:
  if "gmail.com" in i:
    count_gmail+=1
  elif "yahoo.com" in i:
    count_yahoo+=1
  else:
    print("i-->Not a valid Email")
print(f"gmail.com: {round((count_gmail/users_count)*100,2)}%")
print(f"yahoo.com: {round((count_yahoo/users_count)*100,2)}%")

```

**Problem Statement 5: Sales Spike Detection**

**Description:**
A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.

**Requirements:**
Calculate the average daily sales.
Detect days where sales are more than 30% above average.
Display the day number and sale value.

**Input:**
sales = [1200, 1500, 900, 2200, 1400, 3000]

**Expected Output:**
Day 4: 2200
Day 6: 3000

```python
#Taking days input from user, storeing daily sales in list.
#At last printing sales spikes
days = int(input())
sales = []
for i in range(days):
  val = int(input())
  sales.append(val)
avg_sales = sum(sales)/len(sales)
for i in range(len(sales)):
  if sales[i]>avg_sales*0.3+avg_sales:
    print(f"Day {i+1}:{sales[i]}")
```

**Problem Statement 6: Duplicate User ID Detection**

**Description:**
A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.

**Requirements:**
Identify duplicate user IDs.
Display how many times each duplicate appears.

**Input:**
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

**Expected Output:**
user1 → 3 times
user3 → 2 times

```python
#Taking n inputs from user and storeing the result in dictionnary where key is id and value is frequency
#At last printing if frequency of user_ids>1
n = int(input())
user_ids = {}
for i in range(n):
  id = input()
  user_ids[id] = user_ids.get(id,0)+1
for id,count in user_ids.items():
  if count>1:
    print(f"{id}-->{count} times")
```
