**Problem 1:** Smart Light Controller

**Problem Statement**

Create a class to control a smart light that can be turned ON or OFF and display its status.

**Sample Input**

Light Name: Bedroom Light

Action: ON

**Expected Output**

Bedroom Light is ON


```python
#creating a class which handles light name and action
class SmartLight:
  #constructor, self calling while object creation and function calling
  def __init__(self,light_name,action):
    self.light_name = light_name
    self.action = action
    if(self.action == "ON"):
      print(f"{self.light_name} is {self.action}")
    else:
      print(f"{self.light_name} is {self.action}")
#inputs from user
light_name = input("Light Name: ")
action = input("ON/OFF: ")
obj1  = SmartLight(light_name,action)
```

**Problem 2:** Employee ID Card System

**Problem Statement**

Create a class that stores employee name, ID, and department and displays ID card details.

**Sample Input**

 - Employee Name: Rahul
 - Employee ID: EMP102
 - Department: IT

**Expected Output**

Employee ID Card

Name: Rahul

ID: EMP102

Department: IT


```python
#class for storeing employee name, employee id,department
class Id_card:
  def __init__(self,emp_name,emp_id,dept):
    self.emp_name = emp_name
    self.emp_id = emp_id
    self.dept = dept
  #display function
  def display(self):
    print("Employee ID Card")
    print("Name: ",self.emp_name)
    print("ID: ",self.emp_id)
    print("Department: ",self.dept)
#inputs from user
emp_name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")
obj = Id_card(emp_name,emp_id,dept)
obj.display()
```

**Problem 3:** Mobile Contact Record

**Problem Statement**

Create a class to store contact name and phone number and display contact information.

**Sample Input**

Contact Name: Anita

Phone Number: 9876543210

**Expected Output**

Contact Saved

Name: Anita

Phone: 9876543210


```python
#creating class for storeing contact record
class Contact_record:
  def __init__(self,name,phone):
    self.name = name
    self.phone = phone
  #display function
  def display(self):
    print("Contact Saved")
    print("Name: ",self.name)
    print("Phone: ",self.phone)
#input from user
name = input("Contact Name: ")
phone = input("Phone Number: ")
#object creation
obj = Contact_record(name,phone)
obj.display()
```

**Problem 5:** Movie Rating Display System

**Problem Statement**

Create a class that stores movie name and rating and displays movie details.

**Sample Input**

Movie Name: Inception

Rating: 4.8

**Expected Output**

Movie: Inception

Rating: 4.8 / 5


```python
#Movie rating display system
class Movie_rating:
  def __init__(self,name,rating):
    self.name = name
    self.rating = rating
  def display(self):
    print("Movie: ",self.name)
    print(f"Rating: {self.rating} / 5")
#input from user
name = input("Movie Name: ")
rating = float(input("Rating: "))
#object creation
obj1 = Movie_rating(name,rating)
obj1.display()
```

**Problem 6:** Delivery Address Manager

**Problem Statement**

Create a class that stores customer name and delivery address and prints delivery details.


**Sample Input**

Customer Name: Suman

Address: Hyderabad

**Expected Output**

Delivery Details

Customer: Suman

Address: Hyderabad


```python
#Class Delivery address manager which displays and stores name,address
class Delivery_address_Manager:
  def __init__(self,name,address):
    self.name = name
    self.address = address
  def display(self):
    print("Delivery Details")
    print("Customer:",self.name)
    print("Address:",self.address)
#input from user
name = input("Customer Name: ")
address = input("Address: ")
#object creation and calling
obj1 = Delivery_address_Manager(name,address)
obj1.display()
```
