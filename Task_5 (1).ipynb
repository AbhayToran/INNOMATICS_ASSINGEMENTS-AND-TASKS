**Problem 3: Agriculture** – Rainfall Adequacy Checker

**Problem Statement**

Create a function that analyzes seasonal rainfall data.

**Rules**

Calculate average rainfall

Average ≥ required level → Adequate Rainfall

Otherwise → Inadequate Rainfall

**Hint**

Use function

Use loop and math operations

**Expected Output Format**

Average Rainfall: 72

Rainfall Status: Adequate Rainfall

```python
#Here calculating average rainfall
#Returning result if average rainfall is greater than or equall to required level
def status(days,rainfall_data,required_level):
    avg_rainfall = sum(rainfall_data)/days
    if(avg_rainfall >= required_level):
        return "Adequate Rainfall"
    else:
        return "Inadequate Rainfall"
days = int(input())
required_level = int(input())
rainfall_data = []
for i in range(days):
    rainfall = int(input())
    rainfall_data.append(rainfall)
result = status(days,rainfall_data,required_level)
print(f"Average Rainfall: {sum(rainfall_data)/days}")
print(f"Rainfall Status: {result}")
```

**Problem 6: Agriculture** – Premium Crop Price Filter

**Problem Statement**

Create a function that filters crop prices above a premium threshold.

Rules
- Price > ₹2000 → Premium Crop
- Else → Regular Crop

**Hint**

- Use function

- Use list filtering logic

**Expected Output Format**

Premium Crops: [2500, 3200]


```python
#From user taking the crop prices and iterate through the prices if price > 2000 then append in premium group else in regular group
def filter_crop_prices(crop_prices):
  premium_crop = []
  regular_crop = []
  for i in crop_prices:
    if(i > 2000):
      premium_crop.append(i)
    else:
      regular_crop.append(i)
  return premium_crop,regular_crop
n = int(input())
crop_prices = []
for i in range(n):
  price = int(input("Enter Price of Crop: "))
  crop_prices.append(price)
premium_crop,regular_crop = filter_crop_prices(crop_prices)
print(f"Premium Crops: {premium_crop}")
print(f"Regular Crops: {regular_crop}")
```

**Problem 10:** Smart Electricity Bill Analyzer

**Problem Statement**
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
**Billing Rules**
- First 100 units → ₹3 per unit
- Next 100 units (101–200) → ₹5 per unit
- Above 200 units → ₹7 per unit

**Usage Classification**
- Total bill < ₹500 → Low Usage
- ₹500 – ₹1500 → Moderate Usage
- ₹1500 → High Usage

**Hint**

Use function to calculate bill

Use conditional logic for slabs

Use return values (bill and usage status)


```python
#Calculating the usage level and total bill based on set of conditions
def electricity_bill_analyzer(units_consumed):
    total_bill = 0

    if units_consumed <= 100:
        total_bill = units_consumed * 3
    elif units_consumed <= 200:
        total_bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        total_bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)

    usage_level = ""
    if total_bill < 500:
        usage_level = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_level = "Moderate Usage"
    else:
        usage_level = "High Usage"

    return total_bill, usage_level

units = int(input("Enter units consumed: "))
bill, usage_status = electricity_bill_analyzer(units)
print(f"Total Bill: ₹{bill}")
print(f"Usage Level: {usage_status}")
```
